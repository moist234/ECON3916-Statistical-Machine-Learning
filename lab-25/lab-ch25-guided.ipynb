## Part 6: Reflection (10 min)

### Q1: What worked well with AI-assisted development?

The scaffolding speed was the clearest win — a full Next.js project with TypeScript, Tailwind,
dark mode, and five components was generated in a single prompt, something that would have taken
me an hour to wire up manually. Claude Code also handled repetitive token-swapping tasks well,
like replacing every `--accent-blue` reference with `--accent` across six files simultaneously.
Design suggestions were surprisingly coherent: the metric chip on each project card and the
pulsing live dot on the Supabot callout were details I didn't specify but that elevated the UI.
For boilerplate-heavy work — config files, `tsconfig.json`, `.gitignore`, PostCSS — AI removal
of that friction let me stay focused on content rather than setup.

### Q2: What did you have to fix manually?

The biggest issues were environment and dependency errors the model couldn't fully anticipate:
`npx` wasn't on the PATH, `autoprefixer` was missing, and the Google Fonts `@import` was placed
after the `@tailwind` directives — violating CSS spec and causing a 500 error. Claude Code caught
and fixed these iteratively, but each required a feedback loop rather than working first try.
The initial color palette also didn't match my style reference — the warm amber/dark theme had
to be specified precisely in a follow-up prompt because the default output assumed a light
blue-and-white fintech aesthetic. These fixes highlight that AI-generated code is a strong first
draft, not a finished product.

### Q3: How does AI-assisted development compare to writing code entirely yourself?

The HADI loop ran noticeably faster: my hypothesis ("a styled project card with a metric chip")
became a deployed component in minutes rather than hours, which meant I could collect data
(does it look right?) and iterate sooner. However, domain knowledge gaps were real — Claude
had no way to know my actual GitHub URLs, the right metric labels for each project, or that my
style reference implied a warm amber palette rather than a generic teal. The AI handled syntax
and structure; I provided meaning and judgment. Writing code entirely myself would have produced
more intentional decisions at each step but far fewer of them in the same time window.

### Q4: Prompt engineering lessons

Specific, structured prompts dramatically outperform vague ones. My early prompt ("scaffold a
Next.js portfolio site with a hero, projects, skills, and contact section") produced a working
but generic result. My later style prompt — listing exact hex codes, font names by use case, and
pixel values for padding and gutters — produced precise changes with zero back-and-forth.
The lesson: treat Claude Code like a contractor, not a designer. It executes specifications
faithfully but won't infer taste from thin descriptions. A bad prompt says "make it look nicer";
a good one says "shift accent to #F5A623, use Roboto Mono for tags and badges, increase card
padding to 24px."

### Q5: If you were hiring a junior data scientist, would their portfolio site influence your evaluation?

Yes, a portfolio signals things a resume structurally cannot. A resume lists "scikit-learn" as
a skill; a portfolio shows whether the candidate understands *why* PR-AUC matters more than
accuracy on an imbalanced dataset. In labor economics terms, a portfolio is a costly signal:
it takes real time to build and can't be faked at the level of detail a technical evaluator
would probe. It also reveals communication ability — can this person explain a confusion matrix
to a VP of Risk? — which is orthogonal to modeling skill but equally important in practice.
For a junior hire especially, where work history is thin, a well-documented project page
substitutes for experience as evidence of judgment and initiative.
