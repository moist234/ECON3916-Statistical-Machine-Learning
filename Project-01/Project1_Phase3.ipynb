```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

df = pd.read_csv('lalonde.csv')
```

```python
# Step 3.1
# Note: We use np.log() directly in the formula for Log-Level specification
formula_1 = 're78 ~ treat + age + educ + re74 + re75 + black + hispan + married + nodegree'

# Fit the model
model_1 = smf.ols(formula=formula_1, data=df).fit(cov_type='HC1')

# Print the "Regression Anatomy"
print(model_1.summary())

formula_interact = 're78 ~ treat * educ + age + re74 + re75 + black + hispan + married + nodegree'

model_interact = smf.ols(formula=formula_interact, data=df).fit(cov_type='HC1')
print(model_interact.summary())
```

- treat (1548.24, p=0.037): Participation in the job training program is associated with a 1,548 dollar increase in 1978 earnings, ceterus paribus
- educ (403.94, p= 0.012): A one additional year of education is associated with a 404 dollar increase in 1978 earnings, ceteris paribus.
- re74 (0.296, p=0.000): Each additional dollar earned in 1974 is associated with $0.30 more in 1978 earnings, ceteris paribus.
black (-1240.64, p=0.075): Being Black is associated with 1,241 dollars lower earnings in 1978, ceteris paribus, though this falls just short of the p=0.005 significance threshold.
